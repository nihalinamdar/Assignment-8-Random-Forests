```python
import pandas as pd
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
```

```python
fraud_data=pd.read_csv("Fraud_check.csv")
```

```python
fraud_data
```

```python
taxable_income=[]
for value in fraud_data["Taxable.Income"]:
    if value <= 30000:
        taxable_income.append("Risky")
    else:
        taxable_income.append("Good")
        
print(taxable_income)
```

```python
DF2=pd.DataFrame(taxable_income)
DF2=pd.concat([DF2,fraud_data],axis=1)
DF2=DF2.rename({0:'o/p'},axis=1)
```

```python
label_encoder = preprocessing.LabelEncoder()
fraud_data['Undergrad']= label_encoder.fit_transform(fraud_data['Undergrad'])
fraud_data['Marital.Status']= label_encoder.fit_transform(fraud_data['Marital.Status'])
fraud_data['Urban']= label_encoder.fit_transform(fraud_data['Urban'])
DF2['o/p']= label_encoder.fit_transform(DF2['o/p'])
DF2
```

```python
DF2=DF2.values
```

```python
x=DF2[:,1:7]
y=DF2[:,0]
```

```python
num_trees=100
max_features=6
kfold=KFold(n_splits=10) #dataset divided into 10 sets for cross validation
```

```python
model=RandomForestClassifier(n_estimators=num_trees,max_samples=0.8,max_features=max_features,random_state=8)
```

```python
results= cross_val_score(model,x,y,cv=kfold)
print(results.mean()) #mean of the kfold number (10) of results
```
