```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
```

```python
company_data=pd.read_csv("Company_Data.csv")
```

```python
company_data
```

```python
company_data
```

```python
sales_val=[]
for value in company_data["Sales"]:
    if value<=9:
        sales_val.append("low")
    else:
        sales_val.append("high")
        
print(sales_val)
```

```python
df2=pd.DataFrame(sales_val)
```

```python
df3=pd.concat([df2,company_data],axis=1)
df3=df3.rename({0:'o/p'},axis=1)
```

```python
df3=df3.rename({0:'o/p'},axis=1)
```

```python
label_encoder = preprocessing.LabelEncoder()
company_data['ShelveLoc']= label_encoder.fit_transform(company_data['ShelveLoc'])
company_data['Urban']= label_encoder.fit_transform(company_data['Urban'])
company_data['US']= label_encoder.fit_transform(company_data['US'])
df3['o/p']= label_encoder.fit_transform(df3['o/p'])
df3
```

```python
array=df3.values
```

```python
x=array[:,2:10]
y=array[:,0]
```

```python
num_trees=100
max_features=6
kfold=KFold(n_splits=10) #dataset divided into 10 sets for cross validation
```

```python
#0.8 here states that 80 percent of samples are taken under consideration
model=RandomForestClassifier(n_estimators=num_trees,max_samples=0.8,max_features=max_features,random_state=8)
```

```python
results= cross_val_score(model,x,y,cv=kfold) #cross validation tech used here.
print(results.mean()) #mean of the kfold number (10) of results
```
